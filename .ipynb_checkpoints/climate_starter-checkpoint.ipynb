# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
inspector = inspect(engine)
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
#Base.prepare(engine, reflect=True)
```

```python
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
inspector.get_table_names()
```

```python
columns = inspector.get_columns('measurement')
for c in columns:
    print(c['name'], c['type'])
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
recent_date = session.query(Measurement).order_by(Measurement.date.desc()).first()
print(recent_date.date)
```

```python
session.query(func.count(Measurement.date)).all()
```

```python
import datetime as dt
```

```python
# Calculate the date one year from the last date in data set.
twelve_months = dt.date(2017, 8, 23) - dt.timedelta(days=365)
twelve_months
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 

# Perform a query to retrieve the data and precipitation scores
prcp_twelve = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date > twelve_months).all()
print(prcp_twelve)
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
# Sort the dataframe by date
twelve_prcp_df = pd.DataFrame(prcp_twelve, columns = ['date', 'prcp'])
twelve_prcp_df = twelve_prcp_df.set_index('date')
print(twelve_prcp_df)
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
twelve_prcp_df.plot(color='teal')
plt.ylabel('Rain (in)')
plt.xlabel('Dates')
plt.title('Aug 2016-Aug 2017 of Precipitation')
plt.xticks([])
plt.savefig('12 months of PRPC.png')
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
twelve_prcp_df.describe()
```

# Exploratory Station Analysis

```python
columns = inspector.get_columns('station')
for c in columns:
    print(c['name'], c['type'])
```

```python
# Design a query to calculate the total number stations in the dataset
station_total = session.query(Station).group_by(Station.station).count()
print(f'There are {station_total} stations.')
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
active_station = session.query(Measurement.station, func.count(Measurement.date)).group_by(Measurement.station).\
order_by(func.count(Measurement.date).desc()).first()
active_station = active_station.station
print(f'The most active station is {active_station}.')
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
active_station_temp = session.query(func.min(Measurement.tobs), func.max(Measurement.tobs),
func.avg(Measurement.tobs)).\
filter(Measurement.station == active_station).all()
print(f'The Low, High and Average Tempature for {active_station} is {active_station_temp}.')
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
temps = session.query(Measurement.date, Measurement.tobs).\
filter(Measurement.date > twelve_months).\
filter(Measurement.station == active_station).all()
print(temps)
```

```python
temps_df = pd.DataFrame(temps, columns=['date', 'tobs'])
temps_df
```

```python
temps_df.plot(kind='hist', bins=12)
plt.xlabel('Temps in F')
plt.title(f'tobs for {active_station}')
plt.savefig('tobs for USC00519281.png')
plt.show()
```
